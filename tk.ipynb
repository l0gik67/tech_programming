```python
import tkinter as tk                 # импорт основной библиотеки для GUI
from tkinter import messagebox       # импорт модуля для всплывающих окон

# --- функция, которая будет вызываться при нажатии кнопки ---
def on_button_click():
    # получаем текст, который пользователь ввёл в поле Entry
    user_text = entry.get()
    
    # проверяем: пустая строка или нет (strip убирает пробелы)
    if user_text.strip() == "":
        # показываем предупреждающее окно
        messagebox.showwarning("Предупреждение", "Введите текст!")
    else:
        # изменяем текст у label_result (обновление интерфейса)
        label_result.config(text=f"Вы ввели: {user_text}")

# --- создание главного окна приложения ---
root = tk.Tk()                      # создаём объект главного окна
root.title("Пример Tkinter интерфейса")  # задаём заголовок окна
root.geometry("400x250")           # задаём размер окна (ширина x высота)

# --- создание текстовой надписи (Label) ---
label = tk.Label(
    root,                          # родительский контейнер (окно)
    text="Введите текст:",         # текст, который отображается
    font=("Arial", 12)             # шрифт и размер
)
label.pack(pady=10)                # размещаем элемент и добавляем отступы по вертикали

# --- создание поля ввода (Entry) ---
entry = tk.Entry(
    root,                          # родитель
    width=30                       # ширина поля (в символах)
)
entry.pack(pady=5)                 # размещение с небольшим отступом

# --- создание кнопки ---
button = tk.Button(
    root,                          # родитель
    text="Показать",               # текст на кнопке
    command=on_button_click        # функция, вызываемая при нажатии
)
button.pack(pady=10)               # размещение

# --- метка для вывода результата ---
label_result = tk.Label(
    root,
    text="",                       # изначально пустая строка
    font=("Arial", 12),
    fg="blue"                      # цвет текста
)
label_result.pack(pady=10)

# --- запуск главного цикла обработки событий ---
# этот цикл "слушает" действия пользователя (клики, ввод и т.д.)
# без него окно просто закроется сразу после запуска
root.mainloop()
```
